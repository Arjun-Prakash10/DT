### MODEL DEVELOPMENT

```python
x = df.drop(60,axis=1)
y = df.loc[:,60]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train,x_test,y_train,y_test = train_test_split(x, y,test_size=0.3, random_state=7)
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()

kfold = KFold(n_splits=10, random_state=7)
cv_results = cross_val_score(log, x_train_scaled, y_train, cv=kfold, scoring='accuracy')

cv_results.mean()*100
```

```python
from sklearn.tree import DecisionTreeClassifier
dec = DecisionTreeClassifier()

kfold = KFold(n_splits=10, random_state=7)
cv_results = cross_val_score(dec, x_train_scaled, y_train, cv=kfold, scoring='accuracy')

cv_results.mean()*100
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)

kfold = KFold(n_splits=10, random_state=7)
cv_results = cross_val_score(knn, x_train_scaled, y_train, cv=kfold, scoring='accuracy')

cv_results.mean()*100
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()

kfold = KFold(n_splits=10, random_state=7)
cv_results = cross_val_score(nb, x_train_scaled, y_train, cv=kfold, scoring='accuracy')

cv_results.mean()*100
```

```python
from sklearn.svm import SVC
svc = SVC()

kfold = KFold(n_splits=10, random_state=7)
cv_results = cross_val_score(svc, x_train_scaled, y_train, cv=kfold, scoring='accuracy')

cv_results.mean()*100
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()

kfold = KFold(n_splits=10, random_state=7)
cv_results = cross_val_score(rfc, x_train_scaled, y_train, cv=kfold, scoring='accuracy')

cv_results.mean()*100
```

### SVC IS THE MODEL WITH THE BEST ACCURACY

```python
from sklearn.model_selection import GridSearchCV

kfold = KFold(n_splits=10, random_state=7)
param = dict(C = [0.1, 0.5, 1.0, 1.5, 2.0], kernel = ['linear', 'rbf', 'poly'])
grid = GridSearchCV(estimator = svc, param_grid = param, scoring='accuracy', cv=kfold)
grid_result = grid.fit(x_train_scaled, y_train)
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
svc = SVC(C=1.5)
svc.fit(x_train_scaled, y_train)
pred = svc.predict(x_test_scaled)
print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

```python
print(list(y_test))
print(list(pred))
```

```python
df_cnf_matrix = pd.DataFrame(confusion_matrix(y_test, pred),index = ['Mine', 'Rock'], columns = ['Predicted Mine', 'Predicted Rock'])
sns.heatmap(df_cnf_matrix, annot = True)
```
