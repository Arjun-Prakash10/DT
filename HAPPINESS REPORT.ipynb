### FACTORS IMPACT REGION WISE

```python
f,ax = plt.subplots(figsize = (15,7))
sns.set_color_codes("pastel")
sns.barplot(x=df_grp["Economy (GDP per Capita)"],y=df_grp["Region"],color='g',label="Economy")
sns.barplot(x=df_grp["Family"],y=df_grp["Region"],color='b',label="Family")
sns.barplot(x=df_grp["Health (Life Expectancy)"],y=df_grp["Region"],color='c',label="Health")
sns.barplot(x=df_grp["Freedom"],y=df_grp["Region"],color='y',label="Freedom")
sns.barplot(x=df_grp["Trust (Government Corruption)"],y=df_grp["Region"],color='r',label="Trust")
ax.legend(loc="upper right",frameon = True)
ax.set(xlabel='Percentage of Region', ylabel='Region',title = "Factors affecting happiness score")
plt.show()
```

```python
regions = ['Western Europe', 'North America', 'Australia and New Zealand','Middle East and Northern Africa', 'Latin America and Caribbean','Southeastern Asia', 'Central and Eastern Europe', 'Eastern Asia','Sub-Saharan Africa', 'Southern Asia']
plt.figure(figsize=(15,20))
j=0
for i in regions:
    j=j+1
    y = df_2015[df_2015["Region"]==i]['Happiness Score']
    x = df_2015[df_2015["Region"]==i]['Economy (GDP per Capita)']
    plt.subplot(5,2,j)
    ax = sns.regplot(x=x,y=y)
    ax.set(xlabel='Economy (GDP per Capita)', ylabel='Happiness score',title = i)
    plt.tight_layout()
```

Sub-Saharan Africa has the lowest relationship for economy and happiness score

```python
plt.figure(figsize=(15,20))
j=0
for i in regions:
    j=j+1
    y = df_2015[df_2015["Region"]==i]['Happiness Score']
    x = df_2015[df_2015["Region"]==i]['Family']
    plt.subplot(5,2,j)
    ax = sns.regplot(x=x,y=y)
    ax.set(xlabel='Family', ylabel='Happiness score',title = i)
    plt.tight_layout()
```

The happiness score is not affected by increase in family score in Southern Asia and Eastern Asia

```python
plt.figure(figsize=(15,20))
j=0
for i in regions:
    j=j+1
    y = df_2015[df_2015["Region"]==i]['Happiness Score']
    x = df_2015[df_2015["Region"]==i]['Health (Life Expectancy)']
    plt.subplot(5,2,j)
    ax = sns.regplot(x=x,y=y)
    ax.set(xlabel='Health (Life Expectancy)', ylabel='Happiness score',title = i)
    plt.tight_layout()
```

Life expectancy has a strong correlation in most regions and doesn't have much effects in Sub-Saharan Africa, Central and Eastern Europe, Western Europe

```python
plt.figure(figsize=(15,20))
j=0
for i in regions:
    j=j+1
    y = df_2015[df_2015["Region"]==i]['Happiness Score']
    x = df_2015[df_2015["Region"]==i]['Freedom']
    plt.subplot(5,2,j)
    ax = sns.regplot(x=x,y=y)
    ax.set(xlabel='Freedom', ylabel='Happiness score',title = i)
    plt.tight_layout()
```

Freedom in Western Europe and North America is more correlated to happiness score than any other continents.

```python
regions = ['Western Europe', 'North America', 'Australia and New Zealand','Middle East and Northern Africa', 'Latin America and Caribbean','Southeastern Asia', 'Central and Eastern Europe', 'Eastern Asia','Sub-Saharan Africa', 'Southern Asia']
plt.figure(figsize=(15,20))
j=0
for i in regions:
    j=j+1
    y = df_2015[df_2015["Region"]==i]['Happiness Score']
    x = df_2015[df_2015["Region"]==i]['Generosity']
    plt.subplot(5,2,j)
    ax = sns.regplot(x=x,y=y)
    ax.set(xlabel='Generosity', ylabel='Happiness score',title = i)
    plt.tight_layout()
```

Generosity is strongly correlated to happiness score only in Western Europe.

```python
regions = ['Western Europe', 'North America', 'Australia and New Zealand','Middle East and Northern Africa', 'Latin America and Caribbean','Southeastern Asia', 'Central and Eastern Europe', 'Eastern Asia','Sub-Saharan Africa', 'Southern Asia']
plt.figure(figsize=(15,20))
j=0
for i in regions:
    j=j+1
    y = df_2015[df_2015["Region"]==i]['Happiness Score']
    x = df_2015[df_2015["Region"]==i]['Trust (Government Corruption)']
    plt.subplot(5,2,j)
    ax = sns.regplot(x=x,y=y)
    ax.set(xlabel='Trust (Government Corruption)', ylabel='Happiness score',title = i)
    plt.tight_layout()
```

Trust doesn't affect the happiness score in 'Latin America and Caribbean','Southeastern Asia', 'Central and Eastern Europe', 'Eastern Asia','Sub-Saharan Africa' in other regions it has a decent - Strong positive correlation with happiness score
