```python
areas1 = ['Andaman & Nicobar Islands', 'Andhra Pradesh', 'Arunachal Pradesh','Assam', 'Bihar', 'Chandigarh', 'Chhattisgarh']
plt.figure(figsize=(15,10))
plt.xlabel("Years",fontsize = 20)
plt.ylabel("Total No.of Cases",fontsize = 20)
c = ['Red','Blue','Black','Yellow','Brown','Purple','Green']
for i in range(7):
    df_grp_yr = df[df['Area_Name']==areas1[i]].iloc[:,1:].groupby('Year').sum().reset_index()
    sns.lineplot(x = df_grp_yr['Year'],y=df_grp_yr['Victims_of_Rape_Total'],label = areas1[i],color=c[i], linestyle ='-',ci=None)
plt.legend()
         
```
