### FEATURE ENGINEERING

```python
d ={}
for i in df_train["CUISINES"].values:
    i = i.split(",")
    for j in i:
        j = j.strip()
        if j in d:
            d[j]+=1
        else:
            d[j]=1
len(d)
```

```python
cuisines = d.keys()
for i in cuisines:
    df_train[i]=0
    df_test[i]=0
df_train
```

```python
def present(q,i):
    if i in q:
        return 1
    else:
        return 0
```

```python
for i in cuisines:
    df_train[i] = df_train.apply(lambda row:present(row.CUISINES,i), axis = 1)
    df_test[i] = df_test.apply(lambda row:present(row.CUISINES,i), axis = 1)
df_train
```

```python
top_13_cities = df_train['CITY'].value_counts()[:13].index
top_13_cities
```

```python
for i in top_13_cities:
    df_train[i]=0
    df_test[i]=0
df_train
```

```python
df_train['CITY'].fillna('Unknown',inplace=True)
df_test['CITY'].fillna('Unknown',inplace=True)
```

```python
def present_city(q,i):
    if ((i in q) or (q in i)):
        return 1
    else:
        return 0
```

```python
for i in top_13_cities:
    df_train[i] = df_train.apply(lambda row:present_city(row.CITY,i), axis = 1)
    df_test[i] = df_test.apply(lambda row:present_city(row.CITY,i), axis = 1)
df_train
```

```python
d1 ={}
for i in df_train["TITLE"].values:
    i = i.split(",")
    for j in i:
        j = j.strip()
        if j in d:
            d1[j]+=1
        else:
            d1[j]=1
d1
```

```python
title = d1.keys()
for i in title:
    df_train[i]=0
    df_test[i]=0
df_train
```

```python
for i in title:
    df_train[i] = df_train.apply(lambda row:present(row.TITLE,i), axis = 1)
    df_test[i] = df_test.apply(lambda row:present(row.TITLE,i), axis = 1)
df_train
```
