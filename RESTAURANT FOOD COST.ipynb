## RESTAURANT FOOD COST
* TITLE: The feature of the restaurant which can help identify what and for whom it is suitable for.
* RESTAURANT_ID: A unique ID for each restaurant.
* CUISINES: The variety of cuisines that the restaurant offers.
* TIME: The open hours of the restaurant.
* CITY: The city in which the restaurant is located.
* LOCALITY: The locality of the restaurant.
* RATING: The average rating of the restaurant by customers.
* VOTES: The overall votes received by the restaurant.
* COST: The average cost of a two-person meal.

### FEATURE ENGINEERING

##### WE CAN EXTRACT ALL THE DIFFERENT CUISINES FROM THE CUISINES COLUMN AND ASSIGN THEM AS COLS

```python
d ={}
for i in df_train["CUISINES"].values:
    i = i.split(",")
    for j in i:
        j = j.strip()
        if j in d:
            d[j]+=1
        else:
            d[j]=1
len(d)
```

```python
cuisines = d.keys()
for i in cuisines:
    df_train[i]=0
    df_test[i]=0
df_train
```

```python
def present(q,i):
    if i in q:
        return 1
    else:
        return 0
```

```python
for i in cuisines:
    df_train[i] = df_train.apply(lambda row:present(row.CUISINES,i), axis = 1)
    df_test[i] = df_test.apply(lambda row:present(row.CUISINES,i), axis = 1)
df_train
```

### OUTLIER REMOVAL

```python
plt.figure(figsize=(18,10))
cols = ['RATING','VOTES','COST']

for i in range(len(cols)):
    plt.subplot(3,1,i+1)
    ax = sns.boxplot(df_train[cols[i]], orient='h', color = 'green')
    ax.set_xlabel(cols[i],fontsize=15)
```

* There are many outliers in the three columns

```python
temp_df = df_train[cols]
threshold = 3
```

```python
from scipy.stats import zscore
z = np.abs(zscore(temp_df))
np.where(z>threshold)
```

```python
df_new = df_train[(z<threshold).all(axis=1)]
df_new
```

```python
#Percentage of data lost
((len(df_train)-len(df_new))/len(df_train))*100
```

Since loss percentage is more than the threshold(>5%), therefore we will try IQR method

```python
q1 = temp_df.quantile(0.25)
q3 = temp_df.quantile(0.75)
iqr = q3-q1
iqr
```

```python
df_new1 = df_train[~ ( (temp_df<(q1-1.5*iqr)) | (temp_df>(q3+1.5*iqr)) ).any(axis=1)]
df_new1
```

```python
#Percentage of data lost
((len(df_train)-len(df_new1))/len(df_train))*100
```

Since loss percentage is more than the 5%, therefore we will try to increase the threshold in z score method

```python
temp_df = df_train[cols]
threshold = 3.9
```

```python
from scipy.stats import zscore
z = np.abs(zscore(temp_df))
np.where(z>threshold)
```

```python
df_new2 = df_train[(z<threshold).all(axis=1)]
df_new2
```

```python
#Percentage of data lost
((len(df_train)-len(df_new2))/len(df_train))*100
```

## MODEL BUILDING

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_t = sc.fit_transform(x)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error
models = {
    "Linear Regression" : {'model':LinearRegression(),
                           'params':{}
                          }, 
    "Ridge" : {'model':Ridge(),
               'params':{'alpha':[1,2,10,200, 250, 275, 300, 500]}
              }, 
    "Decision Tree" : {'model':DecisionTreeRegressor(),
                       'params':{'max_depth' : [5, 8, 50, 100],
                                 'min_samples_leaf' : [50,100,150,200]
                                }
                      },
    "Random Forest" : {'model':RandomForestRegressor(),
                       'params':{'n_estimators':[100,150],
                                 'max_depth':[5,100,150,200], 
                                 'min_samples_leaf':[1,2,5,8]
                                }
                      },
    "K Nearest Neighbors" : {'model':KNeighborsRegressor(),
                             'params':{'n_neighbors':[5,10,15],
                                       'weights':['uniform','distance'],
                                       'metric':['euclidean','manhattan']
                                      }
                            }
}
```

```python
max_scr = 0

for r_s in range(0,100):
    x_train,x_test,y_train,y_test = train_test_split(x_t,y,random_state=r_s,test_size=0.20)
    for i,j in models.items():
        j['model'].fit(x_train,y_train)
        pred_test = j['model'].predict(x_test)
        r2 = r2_score(y_test,pred_test)
        if r2 > max_scr:
            max_scr = r2
            best_model = j['model']
            best_r_s = r_s
print("Model : ",best_model)
print("Score : ",max_scr)
print("Random State : ",r_s)
```

#### RANDOM FOREST REGRESSOR IS THE BEST MODEL WITH R2 SCORE OF 0.73

```python
from sklearn.model_selection import GridSearchCV
x_train,x_test,y_train,y_test = train_test_split(x_t,y,random_state=9,test_size=0.20)
grid = GridSearchCV(models['Random Forest']['model'], models['Random Forest']['params'], cv=4, scoring = 'r2')
grid.fit(x_train,y_train)
print({
    'model': 'Random Forest',
    'best_score' : np.abs(grid.best_score_),
    'best_params' : grid.best_params_
})
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
rf = RandomForestRegressor(max_depth=100,min_samples_leaf=2,n_estimators=150)
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
print('R2 Score : ',r2_score(y_test,y_pred))
print('RMSE Score : ',np.sqrt(mean_squared_error(y_test,y_pred)))
print('MSE Score : ',mean_squared_error(y_test,y_pred))
print('RMAE Score : ',np.sqrt(mean_absolute_error(y_test,y_pred)))
```
