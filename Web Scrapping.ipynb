 ### 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file. 

```python
page = requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
soup
```

```python
title = soup.find_all('td',class_='titleColumn')
rating_col = soup.find_all('td',class_='ratingColumn imdbRating')

Name = []
Year = []
IMDB_rating = []

for i in range(100):
    Name_col = title[i].text.split()
    Name.append(' '.join(Name_col[1:-1]))
    Year.append(int(title[i].text[-6:-2]))
    IMDB_rating.append(float(rating_col[i].text[1:-1]))
```

```python
import pandas as pd
df = pd.DataFrame({'Name': Name,
     'Year': Year,
     'IMDB_rating': IMDB_rating
    }) 
df
```

```python
df.to_csv('top100_movies_imdb.csv')
```

### 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
page = requests.get("https://bookpage.com/reviews?page=1")
page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
soup
```

```python
title_tag = soup.find_all('h4',class_='italic')
author_tag = soup.find_all('p',class_='sans bold')
genre_tag = soup.find_all('p',class_='genre-links hidden-phone')
# genre_tag
d = {'title' :[], 'author':[], 'genre':[], 'reviews':[]}

for i in  range(5):
    d['title'].append(' '.join(title_tag[i].text.split()[1:]))
    d['author'].append(' '.join(author_tag[i].text.split()))
    d['genre'].append(''.join(genre_tag[i].text.split()))


for link in title_tag[:5]:
    temp = requests.get("https://bookpage.com/"+link.find('a')['href'])
    temp_soup = BeautifulSoup(temp.content, "html.parser")
    x = temp_soup.find('div',class_='article-body')
    d['reviews'].append(x.text)
    print(x)
```

```python
df = pd.DataFrame.from_dict(d) 
df 
```

### 7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
page = requests.get("https://www.amazon.in/s?i=electronics&bbn=1389401031&rh=n%3A976419031%2Cn%3A1389401031%2Cn%3A1389432031%2Cp_36%3A-2000000&dc&qid=1616433861&rnid=1389401031&ref=sr_nr_n_3")
page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
soup
```

```python
div = soup.find_all('div', class_='a-section a-spacing-medium')

d = {'name':[],'price':[],'stars':[], 'img_url' : []}
for item in div:
    name = item.find('span',class_='a-size-base-plus a-color-base a-text-normal')
    star = item.find('span',class_='a-icon-alt')
    price = item.find('span',class_='a-offscreen')
    image_url = item.find('img',class_='s-image')['src']
    
    tmp = name.text
    for i in range(len(tmp)):
        if tmp[i]==')':
            break
    d['name'].append(tmp[:i+1])
    
    
    if star!=None:
        d['stars'].append(float(star.text.split()[0]))
    else:
        d['stars'].append(None)
        
    
    d['price'].append(float(price.text[1:].replace(',','')))
    
    d['img_url'].append(image_url)

d
```

```python
df = pd.DataFrame.from_dict(d) 
df
```

### 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description. 

```python
page = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YFjJZq8zZPY")
page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
soup
```

```python
first = soup.find_all('li',class_='forecast-tombstone')
desc = soup.find_all('div',class_='col-sm-10 forecast-text')

period = []
temp = []
short_desc = []
long_desc = []
for i in range(len(first)):
    tmp = first[i].text.split()
    period.append(tmp[0])
    temp.append(tmp[-2]+tmp[-1])
    short_desc.append(' '.join(tmp[1:-2]))

for i in range(len(first)):
    tmp =' '.join(desc[i].text.split())
    long_desc.append(tmp)
print(len(period), len(temp), len(short_desc), len(long_desc),sep='\n')
```

```python
import pandas as pd
df = pd.DataFrame({'Period': period,
     'temp': temp,
     'short_desc': short_desc,
    'long_desc' : long_desc
    }) 
df
```
