### 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#Fetching Product URL for first 3 pages
url = []

for i in range(3):
    try:
        temp = driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')
        for i in temp:
            url.append(i.get_attribute('href'))
        try:
            next_btn = driver.find_element_by_xpath("//li[@class='a-last']")
            next_btn.click()
            time.sleep(2)
        except NoSuchElementException:
            print("Displayed Results less than three pages")
    except:
        print("Exception")
```

```python
## Initialising a dictionary to store the data
d = {
    'Brand' : [],
    'Product_Name' : [],
    'Model' : [],
    'Ratings' : [],
    'No_of_Ratings' : [],
    'Price' : [],
    'Return_Exchange' : [],
    'Expected_Delivery' : [],
    'Availability' : [],
    'Other_Details' : [],
    'Product_URL' : []
}
```

```python
length = len(url)
length
```

```python
#Fetching "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL” details
j=1
for i in url:
    driver.get(i)
    time.sleep(2)
    
    try:
        d['Brand'].append(driver.find_element_by_xpath("//th[contains(text(),'Brand')]//following::td[1]").text)
    except NoSuchElementException as e:
        d['Brand'].append("-")
        
    try:    
        d['Product_Name'].append(driver.find_element_by_xpath("//span[@id='productTitle']").text)
    except NoSuchElementException as e:
        d['Product_Name'].append("-")
        
    try:
        d['Model'].append(driver.find_element_by_xpath("//th[contains(text(),'Model')]//following::td[1]").text)
    except NoSuchElementException as e:
        d['Model'].append("-")
        
    try:
        d['Ratings'].append(driver.find_element_by_xpath("//span[@data-hook='rating-out-of-text']").text)
    except NoSuchElementException as e:
        d['Ratings'].append("-")
        
    try:
        d['No_of_Ratings'].append(driver.find_element_by_xpath("//div[@data-hook='total-review-count']").text)
    except NoSuchElementException as e:
        d['No_of_Ratings'].append("-")
        
    try:
        d['Price'].append(driver.find_element_by_xpath('//span[contains(@id,"priceblock")]').text)
    except NoSuchElementException as e:
        d['Price'].append("-")
        
    try:
        d['Return_Exchange'].append(driver.find_element_by_xpath('//div[@id="RETURNS_POLICY"]').text)
    except NoSuchElementException as e:
        d['Return_Exchange'].append("-")

    try:    
        d['Expected_Delivery'].append(driver.find_element_by_xpath('//div[@id="ddmDeliveryMessage"]/b').text)
    except NoSuchElementException as e:
        d['Expected_Delivery'].append("-")
        
    try:
        d['Availability'].append(driver.find_element_by_xpath('//div[@id="availability"]').text)
    except NoSuchElementException as e:
        d['Availability'].append("-")
        
    try:    
        d['Other_Details'].append(driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']").text)
    except NoSuchElementException as e:
        d['Other_Details'].append("-")
        
    d['Product_URL'].append(i)
    
    percent = int(j*100/length)
    if( (percent%10) == 0 ):
        print(percent,"%  --------",j,"Out of", length, "complete")
    j+=1
    time.sleep(2)
```

```python
# Converting to DataFrame
df = pd.DataFrame(d)
df
```

### 3. Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.

```python
d={
    'Fruits':[],
    'Cars':[],
    'Machine_Learning':[]
}
```

```python
def load_get_images(search_item):

    # Fetching the website
    driver.get("https://www.google.co.in/imghp?hl=en&authuser=0&ogbl")
    
    # Entering required item to search in search bar
    search_bar = driver.find_element_by_xpath("//input[@class='gLFyf gsfi']")
    search_bar.send_keys(search_item)
    time.sleep(2)
    
    # Clicking Search Button
    search_btn =  driver.find_element_by_xpath("//button[@class='Tg7LZd']")
    search_btn.click()
    time.sleep(3)
    
    # Scrolling Page
    for i in range(2):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)
        
    images = driver.find_elements_by_xpath("//img[@class='rg_i Q4LuWd']")
    
    return images   
```

```python
images = load_get_images('Fruits')
print("\033[1m---------------------------Fruits--------------------------\033[0m")

for i in range(105):
    try:
        images[i].click()
        time.sleep(2)
        tmp = driver.find_elements_by_xpath("//img[@class='n3VNCb']")
        for k in tmp:
             if 'http' in k.get_attribute('src'):
                d['Fruits'].append(k.get_attribute('src'))
        if((i+1)%25==0):
            print(i+1,"out of 100 complete")
    except ElementClickInterceptedException:
        print("Exception : ",i)
        continue
```

```python
images = load_get_images('Cars')
print("\033[1m---------------------------Cars--------------------------\033[0m")

for i in range(105):
    try:
        images[i].click()
        time.sleep(2)
        tmp = driver.find_elements_by_xpath("//img[@class='n3VNCb']")
        for k in tmp:
             if 'http' in k.get_attribute('src'):
                d['Cars'].append(k.get_attribute('src'))
        if((i+1)%25==0):
            print(i+1,"out of 100 complete")
    except ElementClickInterceptedException:
        print("Exception : ",i)
        continue
```

```python
images = load_get_images('Machine Learning')
print("\033[1m---------------------------Machine Learning--------------------------\033[0m")

for i in range(105):
    try:
        images[i].click()
        time.sleep(2)
        tmp = driver.find_elements_by_xpath("//img[@class='n3VNCb']")
        for k in tmp:
             if 'http' in k.get_attribute('src'):
                d['Machine_Learning'].append(k.get_attribute('src'))
        if((i+1)%25==0):
            print(i+1,"out of 100 complete")
    except ElementClickInterceptedException:
        print("Exception : ",i)
        continue
```

### 6. Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 – September 20) from trak.in.

```python
# Fetching the website
driver.get("https://trak.in/")
```

```python
# Clicking Search Button
search_btn =  driver.find_element_by_xpath("//li[@id='menu-item-51510']/a")
search_btn.click()
time.sleep(2)
```

```python
data_list = []
```

```python
for k in ['July', 'August', 'September']:
    option = driver.find_element_by_xpath("//h2[contains(text(),'"+k+", 2020')]//following::div[1]/div[1]/label[1]/select/option[@value=100]")
    option.click()
    time.sleep(1)
    
    str_ = "//h2[contains(text(),'"+k+", 2020')]//following::div[7]/table/tbody/tr"
    l = driver.find_elements_by_xpath(str_)
    
    for i in range(len(l)):
        str1 = str_+"["+str(i+1)+"]/td"
        tmp = driver.find_elements_by_xpath(str1)
        tmp_l = [] 
        for j in tmp:
            tmp_l.append(j.text)
        data_list.append(tmp_l)
```

```python
# Converting to DataFrame
df = pd.DataFrame(data_list, columns =['Sr No.', 'Date(dd/mm/yyyy)', 'Startup Name', 'Industry/Vertical', 'Sub-Vertical', 
                                       'City/Location', "Investor's_Name", 'Investment_Type', 'Amount(In USD)'])
df.drop('Sr No.',axis=1,inplace=True)
df
```
